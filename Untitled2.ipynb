```python

# --- Question 4: Extract GameStop Revenue Data ---
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: URL for GameStop revenue page
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"

# Step 2: Fetch the HTML page with user-agent header
html = requests.get(url, headers={"User-Agent": "Mozilla/5.0"}).text

# Step 3: Parse HTML with BeautifulSoup
soup = BeautifulSoup(html, "html.parser")

# Step 4: Find all tables
tables = soup.find_all("table")

# (Optional) Print table text to locate correct one
# for i, table in enumerate(tables):
#     print(f"Table {i}:\n{table.text[:200]}\n")

# Step 5: Pick correct table (usually table[1] or [2])
revenue_table = tables[1]  # Adjust index if needed

# Step 6: Extract revenue data
data = []
for row in revenue_table.find_all("tr")[1:]:
    cols = row.find_all("td")
    if len(cols) == 2:
        date = cols[0].text.strip()
        revenue = cols[1].text.strip().replace("$", "").replace(",", "")
        if revenue != "":
            data.append([date, float(revenue)])

# Step 7: Convert to DataFrame
gme_revenue = pd.DataFrame(data, columns=["Date", "Revenue (in millions USD)"])

# Step 8: Display last five rows
print(gme_revenue.tail())

```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Step 1: Download Tesla (TSLA) stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

# Step 2: Define the make_graph function
def make_graph(data, title):
    plt.figure(figsize=(12,6))
    plt.plot(data.index, data['Close'], label='Close Price')
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel("Price (USD)")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Step 3: Call the function with Tesla data
make_graph(tesla_data, "Tesla Stock Price Over Time")
```
